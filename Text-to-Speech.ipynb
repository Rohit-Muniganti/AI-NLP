### Text to speech conversion using GTTS

```python
from gtts import gTTS
from IPython.display import Audio

text_to_speech = gTTS('''Welcome to my world, you are currently pursuing full stack datascience and gen ai course.
                         This course will help you to improve your technical skills and coding skills. 
                         If you'd like to share more about your interest and goals which suits you better.
                         Let me know how I can assist you further! Thank you.''',lang='hi',tld='com')

text_to_speech.save('text_to_speech_gtts.wav')
sound_file = 'text_to_speech_gtts.wav'

Audio(sound_file, autoplay = False)
```

### Changing the gender using pyttsx3

```python
import pyttsx3
from IPython.display import Audio

text = '''Welcome to my world, you are currently pursuing full stack datascience and gen ai course.
                         This course will help you to improve your technical skills and coding skills. 
                         If you'd like to share more about your interest and goals which suits you better.
                         Let me know how I can assist you further! Thank you.'''

audio = pyttsx3.init()
audio.setProperty('rate', 150)
audio.setProperty('volume', 0.8)

#Change the voices
voice = audio.getProperty('voices')

#0 for male and 1 for female
audio.setProperty('voice', voice[0].id)   #for female voice

# text to speech conversion
audio.say(text)

#save the audio file
audio.save_to_file(text, 'test_male_Voice.mp3')
#audio.save_to_file(text, 'test_female_Voice.mp3)

audio.runAndWait()
```
